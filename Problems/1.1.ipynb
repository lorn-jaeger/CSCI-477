```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

def sim_a(N, m0, iterations):
    arr = [m0 for _ in range(N)]
    for _ in range(iterations):
        idx1, idx2 = random.sample(range(len(arr)), 2)
        m_i = arr[idx1]
        m_j = arr[idx2]
        
        epsilon = random.uniform(0, 1)
        total = m_i + m_j
        
        arr[idx1] = epsilon * total
        arr[idx2] = (1 - epsilon) * total
        
    return arr

def sim_b(N, m0, iterations, savings):
    arr = [m0 for _ in range(N)]
    for _ in range(iterations):
        idx1, idx2 = random.sample(range(len(arr)), 2)
        m_i = arr[idx1]
        m_j = arr[idx2]
        
        epsilon = random.uniform(0, 1)
        delta_m = (1 - savings) * (epsilon * m_j - (1 - epsilon) * m_i)
        
        arr[idx1] = m_i + delta_m
        arr[idx2] = m_j - delta_m
        
    return arr

def plot(result):    
    plt.figure(figsize=(10, 6))
    sns.histplot(result, bins=30)
    plt.xlabel('Money')
    plt.ylabel('Frequency')
    plt.show()   
```

```python
arr = sim_a(1000, 1000, 1000000)
plot(arr)
```

```python
arr = sim_b(1000, 1000, 1000000, 0.1)
plot(arr)
```

```python
arr = sim_b(1000, 1000, 100000, 0.5)
plot(arr)
```

```python
arr = sim_b(1000, 1000, 100000, 0.75)
plot(arr)
```

```python
arr = sim_b(1000, 1000, 100000, 0.9)
plot(arr)
```

```python
arr = sim_b(1000, 1000, 100000, 0.99)
plot(arr)
```
