```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def sim_a(N, m0, i):
    arr = [m0 for _ in range(N)]

    for _ in range(i):
        idx1, idx2 = random.sample(range(len(arr)), 2)
        m_1 = arr[idx1]
        m_2 = arr[idx2]

        epsilon = random.uniform(0, 1)

        m_1_new = epsilon * (m_1 + m_2)
        m_2_new = (1 - epsilon) * (m_1 + m_2)

        if m_1_new > m_1:  
            transfer = min(m_1_new - m_1, m_2)  
            arr[idx1] += transfer
            arr[idx2] -= transfer
        elif m_2_new > m_2:  
            transfer = min(m_2_new - m_2, m_1)  
            arr[idx1] -= transfer
            arr[idx2] += transfer

    return arr

result = sim_a(100, 1000, 1000)

sns.histplot(result)
```
